```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
# 데이터 읽어오기

data = './Data/가축생산량_쇠고기_1980_2022_FAO.xlsx'

global_beef_production = pd.read_excel(data)


```

```python
global_beef_production
```

```python
# 미국 , 브라질, 중국, 아르헨티나, 호주

global_beef_production.info()
```

```python
global_beef_production.columns
global_beef_production.set_index('국가별',inplace=True)
```

```python
global_beef_production.replace('-', '0', inplace=True)
```

```python
beef_df = global_beef_production.astype('int')
```

```python
beef_df.info()
```

```python
# 파이그래프
plt.figure(figsize=(10,20))
list_beef = []
label_list =  []

for i in beef_df.index :
	
	list_beef.append(beef_df.loc[i].sum())
	label_list.append(i)


# plt.pie(list_beef,labels=label_list)

# plt.legend()
# plt.show()
```

```python


beef_dict = dict()

beef_dict = {label_list[i] : list_beef[i] for i in range(len(label_list))}

print(beef_dict)
```

```python
sorted_beef_dict = dict(sorted(beef_dict.items(), key=lambda item: item[1], reverse=True))

print(sorted_beef_dict)
```

```python
# 키와 값을 리스트로 변환
keys = list(sorted_beef_dict.keys())[1:11]
values = list(sorted_beef_dict.values())[1:11]

plt.figure(figsize=(15,5))
# 그래프 그리기
plt.bar(keys, values)
plt.xlabel('국가별',fontsize=20 )
plt.ylabel('생산량',fontsize=20)
plt.xticks(rotation=45, fontsize=15)
plt.title('국가별 소고기 최대 생산량',fontsize=20)
plt.show()
```

# [미국, 브라질, 중국, 아르헨티나, 오스트레일리아]

```python

file_gases = './Data/세계_온실가스_분야별_배출량_추이_1990_2022.xlsx'

global_data = pd.read_excel(file_gases)
```

```python

global_data.set_index(['국가','배출원'], inplace=True)
```

```python
global_data.columns
```

```python
global_data.drop(columns='항목',inplace=True)
```

```python
global_data.index
```

```python
global_data
```

```python
global_data.loc['미국']
```

```python
df1 = global_data.loc['미국']
df2 = global_data.loc['브라질']
df3 = global_data.loc['중국']
df4 = global_data.loc['아르헨티나']
df5 = global_data.loc['오스트레일리아']
```

```python
df1.columns
```

```python
plt.figure(figsize=(20,5))
plt.suptitle('각 나라의 산업별 온실가스 배출량', fontsize=20)
pie_list = []
pie_name = []

for i in df1.index:
	pie_list.append(int(df1.loc[i].sum()))
	pie_name.append(i)


plt.subplot(1,5,1)
print(pie_list,pie_name)
plt.pie(pie_list, labels=pie_name , autopct='%1.1f%%')
plt.title('미국 산업별 온실가스 배출량',fontsize=12)

pie_list = []
pie_name = []
for i in df2.index:
	pie_list.append(int(df2.loc[i].sum()))
	pie_name.append(i)

plt.subplot(1,5,2)
print(pie_list,pie_name)
plt.pie(pie_list, labels=pie_name , autopct='%1.1f%%')
plt.title('브라질 산업별 온실가스 배출량',fontsize=12)

pie_list = []
pie_name = []
for i in df3.index:
	pie_list.append(int(df3.loc[i].sum()))
	pie_name.append(i)

plt.subplot(1,5,3)
print(pie_list,pie_name)
plt.pie(pie_list, labels=pie_name , autopct='%1.1f%%')
plt.title('중국 산업별 온실가스 배출량',fontsize=12)

pie_list = []
pie_name = []
for i in df4.index:
	pie_list.append(int(df4.loc[i].sum()))
	pie_name.append(i)

plt.subplot(1,5,4)
print(pie_list,pie_name)
plt.pie(pie_list, labels=pie_name , autopct='%1.1f%%')
plt.title('아르헨티나 산업별 온실가스 배출량',fontsize=12)

pie_list = []
pie_name = []
for i in df5.index:
	pie_list.append(int(df5.loc[i].sum()))
	pie_name.append(i)

plt.subplot(1,5,5)
print(pie_list,pie_name)
plt.pie(pie_list, labels=pie_name , autopct='%1.1f%%')
plt.title('오스트레일리아 산업별 온실가스 배출량',fontsize=12)

plt.show()
```
